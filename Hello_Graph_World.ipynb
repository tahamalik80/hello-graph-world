```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Define the nodes
nodes = ["Carol", "Andre", "Fernando", "Diane", "Heather", "Ike", "Jane", "Beverly", "Garth", "Ed"]
G.add_nodes_from(nodes)

# Define the edges based on the given image
edges = [("Carol", "Andre"), ("Carol", "Fernando"), ("Carol", "Diane"), ("Andre", "Diane"), ("Andre", "Fernando"),("Andre", "Beverly"),
         ("Fernando", "Diane"), ("Fernando", "Heather"), ("Diane", "Beverly"), ("Diane", "Garth"),
         ("Beverly", "Ed"), ("Garth", "Ed"),("Garth", "Fernando") , ("Garth", "Beverly"), ("Diane", "Ed"), ("Heather", "Ike"), ("Ike", "Jane"),("Garth", "Heather")]
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(10, 10))
layout = nx.spring_layout(G)
nx.draw(G, pos=layout, with_labels=True, node_color='lightblue', edge_color='purple', node_size=2000, font_size=12, font_weight='bold')
plt.title("Krackhardt Kite Graph")
plt.show()
```
